## Download deals data from BQ

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from google.cloud import bigquery
import os

# Define project on BQ
project_id = "sci-tech-data"

# Authenticate to Google Cloud
client = bigquery.Client(project=project_id)

# Set allow_large_results to True
job_config = bigquery.QueryJobConfig()
job_config.allow_large_results = True

pd.set_option('display.max_columns', None)

os.chdir('/home/jupyter')
!pwd
# # %cd 
# !pwd

%connect_info
# (base) jupyter@starfire-1:~$ pip install jupyter-console
# (base) jupyter@starfire-1:~$ !jupyter console --existing
```

```python
# Download deals

# Download data from BQ
print('downloading pitchbook deals... \n')

# Construct the SQL query to fetch data from BigQuery
query = f'''
SELECT
    *
FROM `starfire-410116.derived.pb_all_deals`
'''

# Fetch data from BigQuery
query_job = client.query(query, job_config=job_config)
results = query_job.result()  # Waits for the query to complete

# Convert the result to a pandas DataFrame
sup = results.to_dataframe()

# Print the first few rows of the DataFrame
print('All deals:')
print(f'shape: {sup.shape}')
sup.head(4)
```

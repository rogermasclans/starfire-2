```python
## Create consecutive lags
def calculate_lag(group):
    # Find the index of the row where Date equals MnAAnnDate
    matching_indices = group.index[group['datadate'] == group['mna_dealdate']].tolist()
    
    if matching_indices:
        # Exact match found
        baseline_index = matching_indices[0]
    else:
        # No exact match, find the first date after MnAAnnDate
        after_mna_date = group[group['datadate'] > group['mna_dealdate'].iloc[0]]
        if not after_mna_date.empty:
            baseline_index = after_mna_date.index[0]
        else:
            # If no date is after MnAAnnDate, fallback to using the latest date
            baseline_index = group.index[-1]
    
    # Calculate the lag by subtracting the baseline_index from the current index
    group['lag'] = group.index - baseline_index
    
    return group


project_id = "starfire-410116"
dataset_id = "derived"
table_id = "stock_prices_pb_2"


# Construct a reference to the target table
table_ref = client.dataset(dataset_id).table(table_id)

# Configure the load job to append data to the existing table
job_config = bigquery.LoadJobConfig(
    write_disposition=bigquery.WriteDisposition.WRITE_APPEND
)


## Iterate over df_deals to get stock price data for days around acquisition

# Suppress warnings -- reinstated at the end of the loop
warnings.filterwarnings("ignore")

# Ensure 'TR.MnAAnnDate' is in datetime format
df['mna_dealdate'] = pd.to_datetime(df['mna_dealdate'])

# Initialize an empty DataFrame to store the results
stock_prices_df = pd.DataFrame()
# Initialize an empty DataFrame to store error information
error_log = pd.DataFrame(columns=['Index', 'mna_investorcikcode', 'Error'])

# Counter to track the number of lines processed
counter = 0

print(f"Loop starts...")

for index, row in df[df['mna_investorcikcode'].notna()  & (df['mna_dealdate'].notna()) ].iterrows():

    cikcode = row['mna_investorcikcode']
    mnaanndate = row['mna_dealdate']
    mnapbdealid = row['mna_dealid']
    # internal_index = row['internal_index']
    
    # Calculate start and end dates (mnaanndate - 7 days and mnaanndate + 7 days)
    start_date = (mnaanndate - timedelta(days=7)).strftime('%Y-%m-%d')
    end_date = (mnaanndate + timedelta(days=7)).strftime('%Y-%m-%d')


    # Fetch historical price summaries
    # try:
    print(f'-----{counter}. Downloading data for {cikcode}...')
    stock_ciks_na = conn.raw_sql(
        f"""
            SELECT
                --gvkey, cik, conm
                datadate, cik, prccd, cshoc, cshtrd
            FROM comp_na_daily_all.secd
            WHERE cik = {cikcode}
            AND {star_date} < datadate AND datadate < {end_date}
        """
    )

    stock_ciks_na = pd.DataFrame(stock_ciks_na)

    # Add additional columns to the result (start, end, MnAAnnDate, RIC)
    stock_ciks_na['Start'] = start_date
    stock_ciks_na['End'] = end_date
    stock_ciks_na['mna_investorcikcode'] = cikcode
    stock_ciks_na['mna_dealdate'] = mnaanndate
    stock_ciks_na['mna_dealid'] = mnapbdealid
    # stock_ciks_na['internal_index'] = internal_index
    stock_ciks_na['loop_counter'] = counter

    print(f'----------stock_ciks_na Shape: {stock_ciks_na.shape}')


    # ## fill missing data from api calls
    # ric_prices[['TR.HQCountryCode', 'TR.ExchangeCountryCode']] = \
    # ric_prices[['TR.HQCountryCode', 'TR.ExchangeCountryCode']].fillna(method='ffill')

    # rename columns to upload to BQ
    # ric_prices = ric_prices.rename(columns=lambda x: x.replace('TR.', '').lower() if x.startswith('TR.') else x)

    # Convert to date format to later process lag functions
    stock_ciks_na['datadate'] = pd.to_datetime(stock_ciks_na['datadate'])


    # stock_ciks_na = stock_ciks_na.groupby('mna_dealid').apply(calculate_lag).reset_index(drop=True)

    # Append the result to the historical_data DataFrame
    stock_prices_df = pd.concat([stock_prices_df, stock_ciks_na], ignore_index=True)
    print(f'----------stock_prices_df Shape: {stock_prices_df.shape}')
    counter += 1

#         # Print progress every 100 lines
#         if counter % 50 == 0:
#             print(f'\n\n\n-----Processed {counter} lines')

#             print(f'----------Appending chunk to BQ...')
#             # Load the DataFrame into BigQuery
#             load_job = client.load_table_from_dataframe(stock_prices_df, table_ref, job_config=job_config)
#             # Wait for the load job to complete
#             load_job.result()
#             print(f'----------Data appended successfully\n\n\n')

#             # Reinitialize master_df to not store duplicate info next time this runs
#             stock_prices_df = pd.DataFrame()


#     except Exception as e:
#         print(f"An error occurred for index: {index}, counter: {counter}, RIC {mna_investorcikcode}: {e}")
#         # Create a DataFrame for the error entry
#         error_entry = pd.DataFrame({'Index': [index], 'CIK code': [mna_investorcikcode], 'Error': [str(e)]})
#         # Concatenate the error entry to the error_log DataFrame
#         error_log = pd.concat([error_log, error_entry], ignore_index=True)

        
# Restore warnings
warnings.filterwarnings("default")

# Display the final historical_data DataFrame
print(stock_prices_df)

# Display the error log DataFrame
print(error_log)
```

```python
############################################################################
##
## Get acquirer CRSP ids (cusip, permno,...) using PB cik
##
############################################################################


# Get clean list of Acquirer PB's cik

acquirer_pb_ciks = df[['mna_investorcikcode']].dropna()
acquirer_pb_ciks['acquirer_pb_cik'] = acquirer_pb_ciks['mna_investorcikcode'].str.split(',')
acquirer_pb_ciks = acquirer_pb_ciks.explode('acquirer_pb_cik').reset_index() #Explode bc some columns contain two ciks
acquirer_pb_ciks['acquirer_pb_cik'] = [item.strip().rjust(10, '0') for item in acquirer_pb_ciks['acquirer_pb_cik']]
print(f'------------------------------------------')
print(f'Total Acquirer PB ciks: {len(acquirer_pb_ciks)}')
print(f'Unique Acquirer PB ciks: {len(acquirer_pb_ciks["acquirer_pb_cik"].unique())}')
print(f'------------------------------------------\n\n')

# Convert list to single string to input to SQL query as a filter
acquirer_pb_ciks_string = ', '.join([f"'{item}'" for item in acquirer_pb_ciks['acquirer_pb_cik'].tolist()])

acquirer_pb_ciks
```

```python
# Get rest of ids from CRSP. CRSP Lookup table contains company ids and industry relationships
acquirer_crsp_data = conn.raw_sql(
f"""
    SELECT 
        cik AS acq_crsp_cik,
        cusip AS acq_cusip,
        gvkey AS acq_gvkey, 
        lpermno AS acq_lpermno, 
        lpermco AS acq_lpermco
    FROM crsp.ccm_lookup
    WHERE cik IN ({acquirer_pb_ciks_string})
"""
)

# Convert to integers while keeping NaN values intact
acquirer_crsp_data['acq_lpermno'] = acquirer_crsp_data['acq_lpermno'].astype('Int64')
acquirer_crsp_data['acq_lpermco'] = acquirer_crsp_data['acq_lpermco'].astype('Int64')
acquirer_crsp_data['acq_lpermno'] = acquirer_crsp_data['acq_lpermno'].astype('object')
acquirer_crsp_data['acq_lpermco'] = acquirer_crsp_data['acq_lpermno'].astype('object')

acquirer_crsp_data = acquirer_crsp_data.drop_duplicates()

print(f'------------------------------------------')
print(f'Acquirer CRSP ids table:\n\n{acquirer_crsp_data}')
print(f'\nUnique Acquirer CRSP cik: {len(acquirer_crsp_data["acq_crsp_cik"].unique())}')
print(f'\nUnique Acquirer CRSP cusip: {len(acquirer_crsp_data["acq_cusip"].unique())}')
print(f'\nUnique Acquirer CRSP gvkey: {len(acquirer_crsp_data["acq_gvkey"].unique())}')
print(f'\nUnique Acquirer CRSP lpermno: {len(acquirer_crsp_data["acq_lpermno"].unique())}')
print(f'\nUnique Acquirer CRSP lpermco: {len(acquirer_crsp_data["acq_lpermco"].unique())}')
print(f'------------------------------------------\n\n')


```

```python

# Function to keep the row with non-NA 'acq_lpermno' and 'acq_lpermco' if they exist
def keep_non_na_rows(group):
    # Check if there are any rows with non-NA 'acq_lpermno' and 'acq_lpermco'
    non_na_rows = group.dropna(subset=['acq_lpermno', 'acq_lpermco'])
    if not non_na_rows.empty:
        return non_na_rows
    else:
        return group

# Apply the function to each group of 'acq_crsp_cik'
acquirer_crsp_data = acquirer_crsp_data.groupby('acq_crsp_cik', group_keys=False).apply(keep_non_na_rows)
acquirer_crsp_data
```

```python
## Merge to PB table to later merge to deals. PB table contains indexs for merging, important because of exploding, etc.

acquirer_pb_ciks = pd.merge(acquirer_pb_ciks, acquirer_crsp_data[['acq_crsp_cik', 'acq_cusip', 'acq_gvkey', 'acq_lpermno', 'acq_lpermco']], left_on='acquirer_pb_cik', right_on = 'acq_crsp_cik', how='left')
acquirer_pb_ciks['acq_cusip_6'] = acquirer_pb_ciks['acq_cusip'].apply(lambda x: str(x)[:6] if pd.notnull(x) else x)


print(acquirer_pb_ciks[acquirer_pb_ciks.duplicated(subset='index', keep=False)])
```

```python
############################################################################
##
## COMPUTE RETURNS
##
############################################################################
# https://wrds-www.wharton.upenn.edu/pages/get-data/center-research-security-prices-crsp/annual-update/stock-security-files/daily-stock-file/
# https://wrds-www.wharton.upenn.edu/pages/get-data/center-research-security-prices-crsp/annual-update/crspcompustat-merged/security-daily/
# https://wrds-www.wharton.upenn.edu/pages/get-data/center-research-security-prices-crsp/annual-update/index-sp-500-indexes/index-file-on-sp-500/
# https://wrds-www.wharton.upenn.edu/pages/support/support-articles/crsp/index-and-deciles/msp500-vs-msp500p-and-dsp500-vs-dsp500/#detail


# Set acquirer ids and event date
ed = '2018-01-09'
acq_cusip = '87621420'
acq_lpermno = '28222'
acq_lpermco = '28222'
lower_date = -10
upper_date = 10

def stock_mkt_returns(ed, acq_cusip, acq_lpermno, acq_lpermco, lower_date, upper_date):
    # Set time window from event date
    ed = pd.Timestamp(ed)
    tw = np.array([ed + pd.DateOffset(days=i) for i in range(lower_date, upper_date)])
    tw = np.array([ts.strftime("'%Y-%m-%d'") for ts in tw])
    tw = ','.join(tw)

    returns_df = conn.raw_sql(
    f"""
        SELECT 
            t1.cusip,
            t1.permno,
            t1.permco,
            t1.date,
            t1.vol,
            t1.numtrd,
            t1.prc,
            t1.openprc,
            t1.shrout,
            t1.ret,
            t1.retx,
            t2.vwretd,
            t2.vwretx,
            t2.sprtrn,
            t1.ret - t2.vwretd AS ab_ret,
            t1.retx - t2.vwretx AS ab_retx,
            t1.ret - t2.sprtrn AS ab_retsp
        FROM crsp.dsf AS t1
        LEFT JOIN (
            SELECT
                caldt,
                vwretd,
                vwretx,
                sprtrn
            FROM crsp_a_indexes.dsp500 
        ) AS t2
        ON t1.date = t2.caldt
        WHERE 
            t1.date IN ({tw}) 
            AND (
                t1.cusip = '{acq_cusip}' 
                OR t1.permno = '{acq_lpermno}' 
                OR t1.permco = '{acq_lpermco}'
            )
        """
    )

    returns_df['date'] = pd.to_datetime(returns_df['date'])
    returns_df['ed'] = pd.to_datetime(ed)
    returns_df['lag'] = (returns_df['date'] - returns_df['ed']).dt.days
    returns_df['acq_cusip'] = acq_cusip
    returns_df['acq_lpermno'] = acq_lpermno
    returns_df['acq_lpermco'] = acq_lpermco
    
    return returns_df
    # Append the new row to the DataFrame

    
columns = ['cusip', 'permno', 'permco', 'date', 'vol', 'numtrd', 'prc', 'openprc', 'shrout', 'ret', 'retx', 'vwretd', 'vwretx', 'sprtrn', 'ab_ret', 'ab_retx', 'ab_retsp', 'ed', 'lag', 'acq_cusip', 'acq_lpermno', 'acq_lpermco']
results_df = pd.DataFrame(columns=columns)

returns_df = stock_mkt_returns(ed, acq_cusip, acq_lpermno, acq_lpermco, lower_date, upper_date)
results_df = pd.concat([results_df, returns_df], ignore_index=True)
results_df


    
```

```python
# Function to rescale lags within each cusip group to account for weekends and other missing days
def rescale_lags(group):
    # Check if lag == 0 is present
    if 0 in group['lag'].values:
        baseline = 0
    else:
        # If lag == 0 isn't present, use lag == 1 as the baseline
        baseline = 1

    # Create negative and positive lag mappings separately
    negative_lags = sorted([lag for lag in group['lag'] if lag < baseline], reverse=True)
    positive_lags = sorted([lag for lag in group['lag'] if lag >= baseline])

    # Create mappings
    neg_mapping = {lag: -index for index, lag in enumerate(negative_lags, start=1)}
    pos_mapping = {lag: index for index, lag in enumerate(positive_lags, start=0)}

    # Merge the two mappings
    lag_mapping = {**neg_mapping, **pos_mapping}

    # Apply the mapping to create the new lag2 column
    group['lag2'] = group['lag'].map(lag_mapping)
    return group



# Drop duplicate firm-lag
result_df = result_df.drop_duplicates(subset=['cusip', 'permno', 'permco', 'lag'])


# Apply the function to each group by cusip
result_df = result_df.groupby('cusip').apply(rescale_lags)

# result_df = result_df[(result_df['lag2'] >= -5) & (result_df['lag2'] <= 5)]

result_df.drop(columns='cusip', inplace=True)

if 'cusip' in result_df.index.names:
    result_df.reset_index(inplace=True)
    
result_df.sort_values(by=['cusip', 'lag2'], inplace=True)

result_df['mna_year'] = result_df['date'].str[:4].astype('Int64')
result_df
```

```python


# Group by 'lag' and calculate the mean of 'ab_ret'
grouped_data = result_df.groupby('lag2')['vol'].mean()

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(grouped_data.index, grouped_data.values, marker='o')
plt.xlabel('Lag')
plt.ylabel('Average ab_ret')
plt.title('Average ab_ret by Lag')
plt.grid(True)
plt.show()
```

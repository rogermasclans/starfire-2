```python
# Download startups

# Download data from BQ
print('downloading pitchbook startups_all... \n')

# Construct the SQL query to fetch data from BigQuery
query = f'''
SELECT
    *
FROM `starfire-410116.derived.startups_all`
'''

# Fetch data from BigQuery
query_job = client.query(query, job_config=job_config)
results = query_job.result()  # Waits for the query to complete

# Convert the result to a pandas DataFrame
sup = results.to_dataframe()

# Print the first few rows of the DataFrame
print('All startups:')
print(f'shape: {sup.shape}')
sup.head(4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and is already loaded with columns 'yearfounded' and 'primaryindustrysector'

# Group by 'yearfounded' and 'primaryindustrysector' and count the occurrences
grouped_data = sup[sup['yearfounded'] >= 1960].groupby(['yearfounded', 'primaryindustrysector']).size().reset_index(name='counts')

# Pivot the data to have years as rows and industry sectors as columns
pivot_data = grouped_data.pivot_table(index='yearfounded', columns='primaryindustrysector', values='counts', fill_value=0)

# Plot the data
pivot_data.plot(kind='line', figsize=(10, 6))

plt.title('Count of Companies by Primary Industry Sector Over Time')
plt.xlabel('Year Founded')
plt.ylabel('Count')
plt.legend(title='Primary Industry Sector')
plt.grid(True)
plt.show()

```

## Investment

```python
# Download startups

# Download data from BQ
print('downloading pitchbook deals_all... \n')

# Construct the SQL query to fetch data from BigQuery
query = f'''
SELECT
    t1.*,
    t2.primaryindustrysector,
    t2.pat_count,
    t2.primaryindustrycode
FROM `starfire-410116.derived.pb_all_deals` t1
LEFT JOIN (
    SELECT
        targetid,
        primaryindustrysector,
        primaryindustrycode,
        pat_count
    FROM `starfire-410116.derived.startups_all`
) AS t2
ON t1.companyid = t2.targetid
'''

# Fetch data from BigQuery
query_job = client.query(query, job_config=job_config)
results = query_job.result()  # Waits for the query to complete

# Convert the result to a pandas DataFrame
deals = results.to_dataframe()

# Print the first few rows of the DataFrame
print('All startups:')
print(f'shape: {deals.shape}')
deals.head(4)
```

```python

# Mapping of PrimaryIndustryCode to industry groups
industry_mapping = {
    # Software
    'Application Software': 'Software',
    'Automation/Workflow Software': 'Software',
    'Business/Productivity Software': 'Software',
    'Communication Software': 'Software',
    'Database Software': 'Software',
    'Educational Software': 'Software',
    'Entertainment Software': 'Software',
    'Financial Software': 'Software',
    'Internet Software': 'Software',
    'Network Management Software': 'Software',
    'Operating Systems Software': 'Software',
    'Social/Platform Software': 'Software',
    'Software Development Applications': 'Software',
    'Vertical Market Software': 'Software',
    
    # Semiconductors & Electronic Components
    'Application Specific Semiconductors': 'Semiconductors & Electronic Components',
    'Electronic Components': 'Semiconductors & Electronic Components',
    'General Purpose Semiconductors': 'Semiconductors & Electronic Components',
    'Production (Semiconductors)': 'Semiconductors & Electronic Components',
    'Computers, Parts and Peripherals': 'Semiconductors & Electronic Components',
    'Electronic Equipment and Instruments': 'Semiconductors & Electronic Components',
    
    # Energy
    'Alternative Energy Equipment': 'Energy',
    'Electric Utilities': 'Energy',
    'Energy Exploration': 'Energy',
    'Energy Infrastructure': 'Energy',
    'Energy Marketing': 'Energy',
    'Energy Production': 'Energy',
    'Energy Refining': 'Energy',
    'Energy Storage': 'Energy',
    'Energy Transportation': 'Energy',

    # Aerospace & Defense
    'Aerospace and Defense': 'Aerospace & Defense',
    
    # Biotechnology & Pharmaceuticals
    'Biotechnology': 'Biotechnology & Pharmaceuticals',
    'Drug Discovery': 'Biotechnology & Pharmaceuticals',
    'Pharmaceuticals': 'Biotechnology & Pharmaceuticals',
    
    # Medical Equipment & Devices
    'Diagnostic Equipment': 'Medical Equipment & Devices',
    'Surgical Devices': 'Medical Equipment & Devices',
    'Therapeutic Devices': 'Medical Equipment & Devices',
    
    # Industrial, Engineering & Equipment
    'Construction and Engineering': 'Industrial, Engineering & Equipment',
    'Industrial Supplies and Parts': 'Industrial, Engineering & Equipment',
    'Machinery (B2B)': 'Industrial, Engineering & Equipment',
    'Oil and Gas Equipment': 'Industrial, Engineering & Equipment',
    'Fiberoptic Equipment': 'Industrial, Engineering & Equipment',
    'Electrical Equipment': 'Industrial, Engineering & Equipment',
    
    # Chemicals & Materials
    'Agricultural Chemicals': 'Chemicals & Materials',
    'Industrial Chemicals': 'Chemicals & Materials',
    'Other Chemicals and Gases': 'Chemicals & Materials',
    'Synthetic Textiles': 'Chemicals & Materials',
    
    # Consumer Goods & Retail
    'Clothing': 'Consumer Goods & Retail',
    'Electronics (B2C)': 'Consumer Goods & Retail',
    'Food Products': 'Consumer Goods & Retail',
    'Internet Retail': 'Consumer Goods & Retail',
}

# Function to map PrimaryIndustryCode to industry
def map_industry(primaryindustrycode):
    return industry_mapping.get(primaryindustrycode)

# Create new column 'industry' based on 'PrimaryIndustryCode'
deals['industry'] = deals['primaryindustrycode'].apply(map_industry)

deals
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and is already loaded with columns 'yearfounded' and 'primaryindustrysector'
df_filtered = deals[deals['dealtype'].isin(['Early Stage VC', 'Later Stage VC'])]
df_filtered['dealyear'] = df_filtered['dealdate'].dt.year
df_filtered['has_patents'] = (df_filtered['pat_count'].fillna(0) > 1).astype(int)

df_filtered = df_filtered[df_filtered['dealyear'] >= 1960]

# Create a new column for 5-year intervals
df_filtered['year_interval'] = (df_filtered['dealyear'] // 5) * 5

# Group by 'year_interval' and 'primaryindustrysector' and sum the 'dealsize'
grouped_data = df_filtered.groupby(['year_interval', 'industry'])['dealsize'].sum().reset_index()

# Pivot the data to have intervals as rows and industry sectors as columns
pivot_data = grouped_data.pivot_table(index='year_interval', columns='industry', values='dealsize', fill_value=0)

# Calculate the relative share by dividing each value by the total for that interval
relative_share = pivot_data.div(pivot_data.sum(axis=1), axis=0)

# Select the top N sectors by their overall share
top_sectors = relative_share.sum().sort_values(ascending=False).head(10).index

# Filter the relative share data to include only the top sectors
relative_share_top = relative_share[top_sectors]

# Plot the relative share data for top sectors
ax = relative_share_top.plot(kind='line', figsize=(10, 6))

# Place the legend outside the plot, centered at the bottom
ax.legend(title='Primary Industry Sector', bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=3)

plt.title('Relative Share of Top 10 Companies by Primary Industry Sector Over 5-Year Intervals (Starting from 1960)')
plt.xlabel('Year Interval')
plt.ylabel('Relative Share')
plt.grid(True)
plt.tight_layout()  # Adjust layout to make space for the legend
plt.show()
```

```python
relative_share_top.plot(kind='area', stacked=True, figsize=(10, 6))
plt.title('Relative Share of Top 10 Companies by Primary Industry Sector (Stacked)')
plt.xlabel('Year Interval')
plt.ylabel('Relative Share')
plt.legend(title='Primary Industry Sector', bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=3)
plt.grid(True)
plt.tight_layout()
plt.show()
```

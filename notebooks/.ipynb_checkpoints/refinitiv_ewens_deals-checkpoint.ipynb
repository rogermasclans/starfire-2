### Download data, connect to refinitiv to get additional data on scd deal id

```python
## Test code

# Download deals
# Define project on BQ
project_id = "sci-tech-data"
# Authenticate to Google Cloud
client = bigquery.Client(project=project_id)

# Set allow_large_results to True
job_config = bigquery.QueryJobConfig()
job_config.allow_large_results = True
# Download data from BQ
print('downloading ewens et al deals... \n')

# Construct the SQL query to fetch data from BigQuery
query = f'''
SELECT
    *
FROM `starfire-410116.derived.dealnum_to_gvkey`
'''

# Fetch data from BigQuery
query_job = client.query(query, job_config=job_config)
results = query_job.result()  # Waits for the query to complete

# Convert the result to a pandas DataFrame
df = results.to_dataframe()

# Print the first few rows of the DataFrame
print('Ewens deals:')
print(f'shape: {df.shape}')
df.head(4)

```

```python
## Download all data for ewens sample

# Initialize an empty DataFrame to store the results
master_df = pd.DataFrame()

# Define the chunk size
chunk_size = 1000
for start in range(0, len(df), chunk_size):
    end = min(start + chunk_size, len(df))
    success = False
    while not success:
        try:
            # Convert the current chunk into a string list that can be passed to the refinitiv's 'screen' function
            mnas_list = ', '.join(f'"{x}"' for x in df['DealNumber'][start:end])

            print(f'Downoading chunk {start}:{end}')
            mna_screen = f'SCREEN(U(IN(DEALS)/*UNV:DEALSMNA*/),\
                                IN(TR.MnASDCDealNumber, {mnas_list}),\
                                CURN=USD)'


            # input ipo screen results as instrument list and pull ipo attributes via fields from Eikon Data API. 
            chunk_df = rdp.get_data(mna_screen, ['TR.MnASDCDealNumber',
                               'TR.MnAAnnDate',
                               'TR.MnARankValueIncNetDebt(Curn=USD,Scale=6)',
                               'TR.MnADealValue(Curn=USD,Scale=6)', 'TR.MnADealCurrency',
                               'TR.MnADealType',
                               'TR.MnAType',
                               'TR.MnATarget', 'TR.MnATargetNation', 'TR.MnATargetSDCCusip',
                               'TR.MnAAcquiror', 'TR.MnAAcquirorNation', 'TR.MnAAcquirorSDCCusip',
                               'TR.MnAAcquirorUltParent', 'TR.MnAAcquirorUltParentSDCCusip',
                               'TR.MnAPctHeldAtAnnDate', 'TR.MnAPctOfSharesAcquired', 'TR.MnAPctOfSharesAcqSeeksToOwn',
                               'TR.MnAStatus',
                               'TR.MnAAcquirorPriTickerSym',
                               'MnAAcquirorUltParentPriTickerSym',
                               'TR.MnACompDealId(Concat="|")']) 

            print(f'Chunk download successful. Shape: {chunk_df.shape}')

            # Append the results to the master DataFrame
            master_df = pd.concat([master_df, chunk_df], ignore_index=True)

            print(f'Master shape: {master_df.shape}')
            print(master_df['TR.MNASDCDealNumber'])

            success = True 
            
        except Exception as e:
            print(f"An error occurred: {e}. Retrying for chunk {start}:{end}...")
    
# Output the master DataFrame
print(master_df)
```

```python
master_df.to_csv('ewens_extended.csv')
```
